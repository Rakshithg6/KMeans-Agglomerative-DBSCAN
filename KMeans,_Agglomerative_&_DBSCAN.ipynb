```python
from sklearn import datasets

#Loading the IRIS dataset
iris = datasets.load_iris()

#Loading only the first two features (sepal length and sepal width) so that clustering could be shown in the 2D space
iris_dataset_features = iris.data[:, :2]
iris_dataset_labels = iris.target

import matplotlib.pyplot as plt

#Plotting a scatter-plot to show initial clustering of the data points
plt.scatter(iris_dataset_features[:,0], iris_dataset_features[:,1], c=iris_dataset_labels, cmap='gist_rainbow')
plt.xlabel('Sepa1 Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)

#Building the KMeans classifier
from sklearn.cluster import KMeans
kmeans_classifier = KMeans(n_clusters = 3,random_state=21)

#Fitting the classifier on the IRIS dataset
kmeans_classifier.fit(iris_dataset_features)

#Getting the model arguments
centers = kmeans_classifier.cluster_centers_
print("Centers: \n", centers, "\n")

new_labels = kmeans_classifier.labels_
print("New labels : \n", new_labels, "\n")

#Plotting the original versus clustered datapoints (different classes are shown in different colors)

fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(iris_dataset_features[:, 0], iris_dataset_features[:, 1], c=iris_dataset_labels, cmap='gist_rainbow',edgecolor='k', s=150)
axes[1].scatter(iris_dataset_features[:, 0], iris_dataset_features[:, 1], c=new_labels, cmap='jet',edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length', fontsize=18)
axes[0].set_ylabel('Sepal width', fontsize=18)
axes[1].set_xlabel('Sepal length', fontsize=18)
axes[1].set_ylabel('Sepal width', fontsize=18)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
```

**DBSCAN**

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize
from sklearn import datasets

#Loading the IRIS dataset
iris = datasets.load_iris()

#Loading only the first two features (sepal length and sepal width) so that clustering could be shown in the 2D space
iris_dataset_features = iris.data[:, :2]
iris_dataset_labels = iris.target

import matplotlib.pyplot as plt
```

```python
db_default = DBSCAN(eps = 0.0375, min_samples = 3).fit(iris_dataset_features)
new_labels = db_default.labels_
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(iris_dataset_features[:, 0], iris_dataset_features[:, 1], c=iris_dataset_labels, cmap='gist_rainbow',edgecolor='k', s=150)
axes[1].scatter(iris_dataset_features[:, 0], iris_dataset_features[:, 1], c=new_labels, cmap='jet',edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length', fontsize=18)
axes[0].set_ylabel('Sepal width', fontsize=18)
axes[1].set_xlabel('Sepal length', fontsize=18)
axes[1].set_ylabel('Sepal width', fontsize=18)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
```
